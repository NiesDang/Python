```python
def percentage(x,o,z):
    if 'AGE2' in x.columns:
        a= x[[(x['AGE2']>= 1)] and (x['AGE2']<=3)]
        m= a[(a[o]<z)]
    else:
        a= x[[(x['AGE3']>= 1)] and (x['AGE3']<=3)]
        m= a[(a[o]<z)]
    sample_percentage = (len(m) / len(x)) * 100
    return sample_percentage
```

```python
def percentage2(x,o,z):
    if 'AGE2' in x.columns:
        a= x[[(x['AGE2']>= 4)] and (x['AGE2']<=6)]
        m= a[(a[o]<z)]
    else:
        a= x[[(x['AGE3']>= 4)] and (x['AGE3']<=6)]
        m= a[(a[o]<z)]
    sample_percentage = (len(m) / len(x)) * 100
    return sample_percentage
```

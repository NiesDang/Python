```python
# there are 2 different function for after 2015 and before 2015. Due to the original resource dont have some specific variables such as methaphetamin before 2015
list1 = ['YESCHFLT',
'YESCHWRK',
'YESCHIMP', 
'YESCHINT' ,
'YETCGJOB',
'YELSTGRD',
'YEPCHKHW',
'YEPHLPHW',
'YEPCHORE',
'YEPLMTTV',
'YEPLMTSN',
'YEPPROUD',
'YEYARGUP',
'YEYFGTSW',
'YEYFGTGP', 
'YEYHGUN', 
'YEYSELL',
'YEYSTOLE',
'YEYATTAK',
'YEVIOPRV',
'YESCHACT',
'YECOMACT',
'YEFAIACT',
'YEOTHACT',
'DSTNRV30',
'DSTNRV30',
'DSTCHR30',
'DSTNGD30',
'IMPSOC',
'IMPSOCM',
'YODPREV']
def readdata(y):
    
    x = pd.read_csv(y,sep='	')
    x= x[[(x['AGE2']> 3)] and (x['AGE2']<=13)]
    x.loc[x['MJEVER'] > 5, 'MJEVER'] = np.nan
    x.loc[x['MJYFU'] > 900, 'MJYFU'] = np.nan
    x.loc[x['HERYFU'] > 900, 'HERYFU'] = np.nan
    x.loc[x['MJMFU'] > 80, 'MJMFU'] = np.nan
    x.loc[x['HERYFU'] > 800, 'HERYFU'] = np.nan
    x.loc[x['HERMFU'] > 80, 'HERMFU'] = np.nan
    x.loc[x['COCEVER'] > 80, 'COCEVER'] = np.nan
    x.loc[x['HEREVER'] > 80, 'HEREVER'] = np.nan
    x.loc[x['METHAMEVR'] > 80, 'METHAMEVR'] = np.nan
    for y in list1:
                x[y]= x[y].apply(lambda a : a if a <80 else np.nan)
                x[y] = x[y].astype('category')

    return x
```
